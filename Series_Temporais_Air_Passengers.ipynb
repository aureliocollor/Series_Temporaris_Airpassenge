Series Temporais Air Passengers:

A análise de séries temporais é um método estatístico que lida com dados de séries temporais para extrair padrões significativos de dados e outras características da tendência dos dados. Os dados do Air Passengers são um exemplo clássico de séries temporais que registram o número total de passageiros de companhias aéreas



Importando bibliotecas:

Neste trecho, estamos importando duas bibliotecas do Python: pandas e matplotlib. A biblioteca pandas é usada para manipulação e análise de dados. Ela fornece estruturas de dados flexíveis que permitem trabalhar com dados relacionais ou rotulados de maneira fácil e intuitiva. A biblioteca matplotlib é usada para criar visualizações estáticas, animadas e interativas em Python.


```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
```

Importando dados:

Aqui, estamos definindo a URL dos dados que queremos analisar. Neste caso, estamos usando um conjunto de dados disponível publicamente na internet

```python
## URL dos dados
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
```

Lendo os dados:
Usamos a função read_csv do pandas para ler o arquivo CSV disponível na URL especificada. O resultado é armazenado em um DataFrame do pandas, que é uma estrutura de dados bidimensional, semelhante a uma planilha.

```python
# Lendo os dados
dados = pd.read_csv(url)
```

Datatime:

Aqui, vou convertendo a coluna ‘Month’ para o tipo de dados datetime. Isso é útil porque permite que o pandas entenda que essa coluna contém informações de data e hora, o que é essencial ao trabalhar com séries temporais.

```python
# Convertendo a coluna 'Month' para datetime
dados['Month'] = pd.to_datetime(dados['Month'])

# Definindo 'Month' como índice
dados = dados.set_index('Month')
```

Visualizando dados:

Por fim, usando a biblioteca matplotlib para visualizar as série temporal. Criando um gráfico de linha com o tempo no eixo x e o número de passageiros no eixo y. A função show é usada para exibir o gráfico.

```python
# Plotando a série temporal
plt.figure(figsize=(10,6))
plt.plot(dados.index, dados['Passengers'])
plt.title('Número de Passageiros ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Número de Passageiros')
plt.show()
```
